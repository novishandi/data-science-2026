```python
# =========================================================
# 1. IMPORT LIBRARY
# =========================================================

import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import matplotlib.gridspec as gridspec
import seaborn as sns

import ssl
ssl._create_default_https_context = ssl._create_unverified_context

# =========================================================
# 2. LOAD DATASET
# =========================================================

df = sns.load_dataset('tips')

# Menampilkan data awal
print(df.head())

# Menampilkan jumlah data dan tipe data
print(df.shape)
print(df.dtypes)

# Statistik deskriptif
print(df.describe())

# =========================================================
# 3. SET STYLE
# =========================================================

sns.set_theme(style='whitegrid')
```

```python
# =========================================================
# 4. MEMBUAT LAYOUT DASHBOARD
# =========================================================

fig = plt.figure(figsize=(16,12))

fig.suptitle(
    'Dashboard Analisis Dataset Tips',
    fontsize=18,
    fontweight='bold'
)

# Layout 2x2
gs = gridspec.GridSpec(
    2, 2,
    figure=fig,
    hspace=0.4,
    wspace=0.3
)

# Membuat area grafik
ax1 = fig.add_subplot(gs[0,0])
ax2 = fig.add_subplot(gs[0,1])
ax3 = fig.add_subplot(gs[1,0])
ax4 = fig.add_subplot(gs[1,1])
```

```python
# =========================================================
# GRAFIK 1 - BAR CHART
# =========================================================

avg_by_day = df.groupby(
    'day',
    observed=False
)['total_bill'].mean()

avg_by_day.plot(
    kind='bar',
    ax=ax1,
    color='skyblue',
    edgecolor='black'
)

ax1.set_title('Rata-rata Total Bill per Hari')
ax1.set_xlabel('Hari')
ax1.set_ylabel('Total Bill (USD)')
```

### What?
Grafik menunjukkan rata-rata total tagihan pelanggan berdasarkan hari. Hari Sabtu dan Minggu memiliki rata-rata total tagihan yang lebih tinggi dibandingkan hari lainnya.

### So what?
Hal ini menunjukkan bahwa pelanggan cenderung melakukan pengeluaran lebih besar pada akhir pekan. Kemungkinan pelanggan datang bersama keluarga atau teman sehingga jumlah pesanan meningkat.

### Now what?
Restoran dapat menambah persediaan makanan dan jumlah staf pada akhir pekan agar pelayanan tetap optimal saat jumlah pelanggan meningkat.

```python
# =========================================================
# GRAFIK 2 - HISTOGRAM
# =========================================================

sns.histplot(
    data=df,
    x='total_bill',
    kde=True,
    color='orange',
    ax=ax2
)

ax2.set_title('Distribusi Total Bill')
ax2.set_xlabel('Total Bill (USD)')
ax2.set_ylabel('Jumlah Pelanggan')
```

### What?
Histogram menunjukkan distribusi total tagihan pelanggan. Sebagian besar pelanggan memiliki total tagihan pada kisaran 10 hingga 20 USD.

### So what?
Hal ini menunjukkan bahwa mayoritas pelanggan melakukan transaksi dalam jumlah sedang dan hanya sedikit pelanggan dengan tagihan sangat besar.

### Now what?
Restoran dapat membuat promo khusus untuk pelanggan dengan transaksi menengah agar nilai pembelian pelanggan dapat meningkat.

```python
# =========================================================
# GRAFIK 3 - BOXPLOT
# =========================================================

sns.boxplot(
    data=df,
    x='day',
    y='total_bill',
    hue='day',
    palette='Set2',
    legend=False,
    ax=ax3
)

ax3.set_title('Boxplot Total Bill per Hari')
ax3.set_xlabel('Hari')
ax3.set_ylabel('Total Bill (USD)')
```

### What?
Boxplot menunjukkan persebaran total tagihan pada setiap hari. Terdapat beberapa nilai outlier dengan total tagihan yang jauh lebih tinggi dibanding pelanggan lainnya.

### So what?
Hal ini menunjukkan adanya pelanggan tertentu yang melakukan transaksi dalam jumlah besar sehingga mempengaruhi distribusi data.

### Now what?
Restoran dapat mempelajari karakteristik pelanggan dengan transaksi tinggi untuk memberikan layanan premium atau promo khusus.

```python
# =========================================================
# GRAFIK 4 - SCATTER PLOT
# =========================================================

sns.scatterplot(
    data=df,
    x='total_bill',
    y='tip',
    hue='time',
    palette='Set1',
    ax=ax4
)

ax4.set_title('Hubungan Total Bill dan Tip')
ax4.set_xlabel('Total Bill (USD)')
ax4.set_ylabel('Tip (USD)')
```

### What?
Scatter plot menunjukkan hubungan antara total tagihan dan jumlah tip pelanggan. Semakin besar total tagihan, semakin besar pula tip yang diberikan pelanggan.

### So what?
Hal ini menunjukkan adanya hubungan positif antara total pembelian dan jumlah tip yang diberikan pelanggan kepada pelayan.

### Now what?
Restoran dapat meningkatkan kualitas pelayanan agar pelanggan tetap merasa puas dan memberikan tip yang lebih besar.

```python
# =========================================================
# MENAMPILKAN DAN MENYIMPAN DASHBOARD
# =========================================================

plt.subplots_adjust(
    hspace=0.4,
    wspace=0.3
)

plt.savefig(
    'dashboard_tips.png',
    dpi=150,
    bbox_inches='tight'
)

plt.show()

print("Dashboard berhasil disimpan sebagai dashboard_tips.png")
```

Berdasarkan dashboard visualisasi dataset tips, pelanggan cenderung memiliki total tagihan lebih tinggi pada akhir pekan. Selain itu terdapat hubungan positif antara total tagihan dan jumlah tip pelanggan. Distribusi data juga menunjukkan sebagian besar pelanggan memiliki transaksi dalam kategori menengah dengan beberapa transaksi bernilai tinggi sebagai outlier.
