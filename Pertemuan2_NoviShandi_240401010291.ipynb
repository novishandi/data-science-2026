## Q2: Missing Values

Missing values adalah data yang kosong atau tidak terisi.

Langkah yang dilakukan:
- Menggunakan `isnull()` untuk mendeteksi data kosong
- Menggunakan `sum()` untuk menghitung jumlahnya
- Menghitung persentase missing terhadap total data

Hasil ini penting karena:
- Data kosong dapat mempengaruhi analisis
- Perlu ditangani (diisi atau dihapus) sebelum analisis lanjutan

Kolom dengan missing values terbanyak adalah Cabin, diikuti oleh Age dan Embarked.

- Cabin memiliki banyak data kosong karena tidak semua penumpang tercatat nomor kabinnya
- Age kosong kemungkinan karena data tidak lengkap saat pencatatan
- Embarked hanya sedikit missing sehingga tidak terlalu berpengaruh

Kesimpulan:
Perlu penanganan missing values sebelum analisis lebih lanjut, terutama pada kolom Age dan Cabin.

```python
missing = df.isnull().sum()

# persentase
persen = (missing / len(df)) * 100

# gabungkan
hasil_missing = pd.DataFrame({
    'Jumlah Missing': missing,
    'Persentase (%)': persen
})

print(hasil_missing)
```

## Q3: Persentase Penumpang Selamat

Kolom `Survived` berisi:
- 1 = selamat
- 0 = tidak selamat

Untuk menghitung persentase:
- Digunakan rata-rata (`mean()`)
- Karena nilai hanya 0 dan 1, maka mean langsung merepresentasikan persentase

Hasil kemudian dikalikan 100 agar menjadi persen (%)

```python
persen_selamat = df['Survived'].mean() * 100
print(f"{persen_selamat:.2f}%")
```

## Q4: Wanita Kelas 1

Dilakukan filtering data dengan dua kondisi:
- Jenis kelamin = perempuan (`female`)
- Kelas penumpang = 1

Setelah data difilter:
- Dihitung jumlah penumpang
- Dihitung rata-rata usia

Tujuan:
Mengetahui karakteristik penumpang wanita kelas atas pada Titanic

Jumlah penumpang wanita kelas 1 cukup banyak dan rata-rata usianya berada di kisaran dewasa.

Hal ini menunjukkan bahwa penumpang kelas atas didominasi oleh orang dewasa dengan kondisi ekonomi lebih baik.

```python
wanita_kelas1 = df[(df['Sex'] == 'female') & (df['Pclass'] == 1)]

# jumlah
jumlah = wanita_kelas1.shape[0]

# rata-rata umur
rata_umur = wanita_kelas1['Age'].mean()

print("Jumlah:", jumlah)
print("Rata-rata umur:", rata_umur)
```

## Q5: Tingkat Keselamatan per Kelas

Data dikelompokkan berdasarkan kelas penumpang (`Pclass`) menggunakan `groupby()`.

Kemudian:
- Menghitung rata-rata `Survived` untuk setiap kelas

Hasil:
- Menunjukkan tingkat keselamatan tiap kelas
- Nilai lebih tinggi berarti peluang selamat lebih besar

Digunakan `idxmax()` untuk mengetahui kelas dengan tingkat keselamatan tertinggi.

Berdasarkan hasil perhitungan, kelas 1 memiliki tingkat keselamatan paling tinggi dibanding kelas lainnya.

Hal ini menunjukkan bahwa penumpang kelas atas memiliki peluang selamat lebih besar, kemungkinan karena prioritas evakuasi dan akses yang lebih baik.

```python
group = df.groupby('Pclass')['Survived'].mean()

print(group)

kelas_tertinggi = group.idxmax()
print("Kelas dengan survival tertinggi:", kelas_tertinggi)
```

## Q6: Analisis Tambahan

Analisis dilakukan untuk melihat hubungan antara kelas penumpang (Pclass) dan jenis kelamin (Sex) terhadap tingkat keselamatan.

```python
df.groupby(['Pclass','Sex'])['Survived'].mean()
```

Hasil analisis menunjukkan bahwa tingkat keselamatan berbeda berdasarkan kombinasi kelas dan jenis kelamin.

- Wanita di kelas 1 memiliki tingkat keselamatan paling tinggi
- Pria di kelas 3 memiliki tingkat keselamatan paling rendah

Interpretasi:
Penumpang dengan kelas lebih tinggi dan berjenis kelamin wanita memiliki peluang selamat lebih besar. Hal ini kemungkinan disebabkan oleh prioritas evakuasi serta akses yang lebih baik pada penumpang kelas atas.

Kesimpulan:

Analisis ini menunjukkan bahwa faktor sosial (kelas) dan gender secara bersamaan sangat mempengaruhi peluang keselamatan.
